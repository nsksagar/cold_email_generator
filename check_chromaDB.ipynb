```python
import chromadb
client = chromadb.Client()
```

```python
collection = client.get_or_create_collection(name='my_collection')
```

```python
collection.add(
    ids = ['id1', 'id2'],
    documents = [
        'This is a doc about aeroplanes',
        'This is a doc about fighter jets'
    ]
)
```

```python
all_docs = collection.get()

all_docs
```

```python
docs = collection.get(ids = ['id1'])
docs
```

```python
results = collection.query(
    query_texts = ['Query is about wars'],
    n_results = 2
)
results
```

```python
collection.delete(ids = all_docs['ids'])
collection.get()
```

```python
collection.add(
    documents = [
        'This document is about aeroplanes',
        'This document is about fighter jets'
    ],
    ids = ['id3', 'id4'],
    metadatas = [
        {'url':'https://en.wikipedia.org/wiki/Airplane'},
        {'url':'https://en.wikipedia.org/wiki/Fighter_aircraft'}
    ]
)
```

```python
results = collection.query(
    query_texts = ['Query is about wars'],
    n_results = 2
)
results
```
